# Bird

<a id='bird'></a>

Default import

```python
b = Bird()
b.Plot3D(plot_trail=True)
```

Advanced import

```python
# a more complex load using all the parameters
b = Bird(
    initial_position = [0,0,-0.3] # initial position 
    ,velocity=[0.2,0.3,0.2] # Linear velocity vector

    # Bird Dimensions
    ,body_length = 0.4
    ,wing_len_1 = 0.3  # Length of segment 1
    ,wing_len_2 = 0.3  # Length of segment 2

    # flapping speed and amplitude
    ,A1 = 40  # Amplitude of flapping angle in degrees
    ,A2 = 30  # Amplitude of segment2 flapping angle
    ,ksi10 = 15  # Lag flapping angle in degrees
    ,ksi20 = 40  # Lag flapping angle in degrees
    ,C2 = 20  # Amplitude of segment2 twisting angle
    ,F0 = 1  # Flapping frequency
    )
b.Plot3D(plot_trail=True)
```

# Drone

<a id='drone'></a>

Default import

```python
D = Drone()
D.Plot3D()
```

Advanced import

```python
D = Drone(
    N_blades = 6,  # number of blades (2 or 3)
    blade_length = 4,  # blade length (m)
    blade_width = 0.7,  # blade width (m)
    blade_root = 0.5,  # distance from the center to start the blade (m)
    omega = 12, # blade rotation rate
)
D.Plot3D()
```

# Radar

<a id='radar'></a>

Loading the radar object (showing default parameters)

```python
R = Radar(
    T = 10, # time duration
    n_samples = 8192, # number of samples
    lambda_ = 0.03,  # Wavelength in meters
    rangeres = 0.05,  # Range resolution in meters
    radarloc = np.array([20, 0, -10]),  # Radar location
)
```

# Simulations

<a id='simulate'></a>

Loading the radar object (showing default parameters)

```python
# Lets again load a radar if you haven't
R = Radar()

# and a bird
B = Bird()

# run bird simulation
R.simulate(B)

# plot range profiles
R._plot_range_profile()

# plot micro doppler
R._plot_micro_doppler_signature()
```

### lets simulate two birds

```python
# load birds
B = Bird()
B2 = Bird(
    initial_position = [0,0,-0.3],
    velocity=[0.2,0.3,0.2], 
    wing_len_1=0.3,
    wing_len_2 = 0.3
    )


# load radar
R = Radar()

# simulate birds
R.simulate(B)
R.simulate(B2)

# R._plot_range_profile()
R._plot_micro_doppler_signature()
```

### Small drone and a bird

```python
# load drone
D = Drone(N_blades = 2,  # number of blades (2 or 3)
        blade_length = 0.12,  # blade length (m)
        blade_width = 0.015,  # blade width (m)
        blade_root = 0.007,  # distance from the center to start the blade (m)
)

# load bird
B = Bird()

# load radar
R = Radar()

# simulate birds
R.simulate(B)
R.simulate(B2)

# R._plot_range_profile()
R._plot_micro_doppler_signature()
```
